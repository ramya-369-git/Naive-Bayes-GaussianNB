```python
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target

print(df.head())
```

```python
import matplotlib.pyplot as plt
import numpy as np

for col in iris.feature_names:
    data = df[col]
    mean = data.mean()
    std = data.std()

    plt.figure(figsize=(6, 4))
    plt.hist(data, bins=20, density=True, edgecolor='black', alpha=0.6, label='Data')

    x = np.linspace(data.min(),data.max(),100)
    normal_curve = (1 / (std * np.sqrt(2 * np.pi))) * np.exp(- (x - mean)**2 / (2 * std**2))
    plt.plot(x, normal_curve, 'r--', label='Normal curve')

    plt.title(f'{col} (mean={mean:.2f}, std={std:.2f})')
    plt.xlabel(col)
    plt.ylabel('Density')
    plt.legend()
    plt.tight_layout()
    plt.show()

```

```python
x=df.iloc[:,:-1]
x.head()
```

```python
y=df.iloc[:,-1]
y.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model=GaussianNB()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```
